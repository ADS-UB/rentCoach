```python
import re
text = adresses.values.tolist()
i=0
Barri=[]
Districte=[]
pat_dis = r'.*, (.*), Barcelona.*'         
pat_barri = r'.*, (.*), .*, Barcelona.*' 
for line in text:
    try:
        match = re.search(pat_dis, line)
        Districte.append(match.group(1))
        match2 = re.search(pat_barri, line)
        Barri.append(match2.group(1))
    except Exception:
        Districte.append(np.nan)
        Barri.append(np.nan)
```
