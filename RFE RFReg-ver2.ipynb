```python
%matplotlib inline

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.datasets import make_classification
from sklearn.cross_validation import train_test_split
import pandas as pd
from sklearn import metrics
import numpy as np

# Build a classification task using 3 informative features
data = pd.read_csv('version4_district+neighborhood.csv')
data.head()



```

```python
#Remove outliers


per=np.percentile(np.array(data['price']), [2.5, 89])
data=data[data['price']>per[0]]
data=data[data['price']<per[1]]

data['price'].hist()
```

```python
X=np.array(data.drop(['price'], axis=1))
y=np.array(data['price'])
n_features=X.shape[1]
print n_features

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=44)
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
Facc=np.zeros(n_features)
# Create the RFE object and compute a cross-validated score.
rfr = RandomForestRegressor(n_estimators=750, max_features='sqrt', bootstrap=False, random_state=44, n_jobs=-1)
scaler = StandardScaler()
# The "accuracy" scoring is proportional to the number of correct
# classifications
for i in range(n_features):
    print i
    rfe = RFE(rfr,i+1)
    rfe.fit(X_train,y_train)
    y_hat=rfe.predict(X_test)
    Facc[i]=metrics.mean_absolute_error(y_test, y_hat)
   

print Facc    
print("Optimal number of features : %d" % Facc.argmin())
```

```python
plt.figure()
plt.plot(Facc,'r')
plt.show
```
