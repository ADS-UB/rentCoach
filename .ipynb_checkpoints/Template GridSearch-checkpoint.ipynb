```python
from __future__ import division
import pandas as pd
import numpy as np
data = pd.read_csv('version1.csv')
```

```python
data.columns
```

```python
data = data.drop(['Unnamed: 0','propertyCode','thumbnail','url','district','neighborhood','propertyType','province','showAddress','hasVideo','priceByArea','newDevelopment','numPhotos'], axis=1)
```

```python
data1=data[data['municipality']=='Barcelona']
```

```python
data = data.drop(['municipality'], axis=1)
```

```python
data
```

```python
data[data['floor']==1000]=0
```

```python
data
```

```python
y =data['price']
X = data.drop(['price'], axis=1)
X=np.array(X)
y=np.array(y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn import cross_validation
from sklearn import metrics
from sklearn import grid_search
from sklearn.preprocessing import StandardScaler
```

```python
parameters = {'normalize' : [True,False]}

kf = cross_validation.KFold(n=data.shape[0], n_folds=5, shuffle=False, random_state=0)

yhat = np.empty_like(y)
acc = np.empty(5)
i = 0

for train_index, test_index in kf:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    #scaler = StandardScaler()
    #X_train = scaler.fit_transform(X_train)
    
    #Standard parameters
    clf= LinearRegression(fit_intercept=True)
    # We can change the scoring "average_precision", "recall", "f1"
    clf = grid_search.GridSearchCV(clf, parameters, scoring='mean_absolute_error', n_jobs=1,cv=10,random_state=0)
    clf.fit(X_train,y_train.ravel())
    
    #X_test = scaler.transform(X_test)
    yhat[test_index] = clf.predict(X_test)
    
    # Recall, f1, precision
    acc[i] = metrics.mean_absolute_error(yhat[test_index], y_test)
    print str(clf.best_params_), acc[i]
    i=i+1
    
print 'Mean accuracy: '+ str(np.mean(acc))
```

```python
#Simple train test split with the optimal parameters

X_train,X_test,y_train,y_test = cross_validation.train_test_split(X, y, test_size=0.1, random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
    
clf= LinearRegression(fit_intercept=True,normalize=False)
clf.fit(X_train, y_train.ravel())

X_test = scaler.transform(X_test)
yhat = clf.predict(X_test)
#print metrics.accuracy_score(yhat, y_test)
print metrics.mean_absolute_error(yhat, y_test)
```

```python
X_train,X_test,y_train,y_test = cross_validation.train_test_split(X, y, test_size=0.1, random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
    
clf = svm.SVC(C=10,kernel='rbf', class_weight={1:1}, gamma=1/40)
clf.fit(X_train, y_train.ravel())

X_test = scaler.transform(X_test)
yhat = clf.predict(X_test)
print metrics.accuracy_score(yhat, y_test)
```
